### Parsing do date_time

```python
def parse_timestamp(df, datetime):
    df[datetime] = pd.to_datetime(df[datetime])
    df['Year'] = df[datetime].dt.year
    df['Month'] = df[datetime].dt.month
    df['Weekday'] = df[datetime].dt.weekday
    df['Hour'] = df[datetime].dt.hour

def categorize_hour(hour):    
    cat = 'None'
    
    if hour in [1,2,3,4,5]:
        cat = 'dawn'
    elif hour in [6,7,8,9,10,11,12]:
        cat = 'morning'
    elif hour in [13,14,15,16,17,18]:
        cat = 'afternoon'
    elif hour in [19,20,21,22,23,24]:
        cat = 'night'
        
    return cat
```

Além de categorizarmos os dados consertarei um problema que identifiquei, em alguns objetos existem marcas temporais repetidas e com atributos diferentes. Isso não faz sentido no contexto de nosso dataset, por isso irei removê-los. Mas primeiro vou elimininar as duplicatas existentes.

```python
df.drop_duplicates(inplace = True)

df.duplicated().sum()
```

```python
issue_index = df[df['date_time'].duplicated()].index

df.drop(index = issue_index,inplace = True)
```

# Importância de Atributos

Agora irei calcular a importância dos atributos que serão utilizados.

```python
#df.drop(columns = ['date_time'],inplace = True)
df.head()
```

```python
le = LabelEncoder()
df_num = df.copy()

for column in df_num.columns:
    if df_num[column].dtype == 'object':
        df_num[column] = le.fit_transform(df_num[column])

df_num.head()
```

```python
df_num.columns
```

```python
feats = df_num.drop('traffic_volume',1)
label = df_num['traffic_volume']

feats.head()
```

```python
onehot_feats = feats.copy()
onehot_feats[['holiday','weather_main','Year','Month','Hour','Weekday']] = feats[['holiday','weather_main','Year','Month','Hour','Weekday']].astype('category')
onehot_feats = pd.get_dummies(onehot_feats)

onehot_feats.head()
```

```python
model = RandomForestRegressor(random_state=1, max_depth=10)
model.fit(feats,label)
#Plot the feature importance to see the important features
features = feats.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-4:] #Top 4 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
model = RandomForestRegressor(random_state=1, max_depth=10)
model.fit(onehot_feats,label)
#Plot the feature importance to see the important features
features = onehot_feats.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-10:] #Top 10features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

# Treino e Teste

## Train-Test SPLIT
### Train = 0.7; Test = 0.3

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train.shape
```

### Random Forest

```python
reg = RandomForestRegressor()
reg = reg.fit(X_train,y_train)
predict = reg.predict(X_test)
accuracy = reg.score(X_test, y_test)
mae = mean_absolute_error(y_test, predict)

print(f"R2 Score: {round(accuracy * 100, 3)}%")
print(f"Mean Absolute Error: {round(mae, 4)}")
```

```python
print(f'RMSE: {round(np.sqrt(mean_squared_error(y_pred=predict,y_true=y_test)),4)}')
```

## XGBoost

```python
xgb_model = xgb.XGBRegressor(objective="reg:squarederror",  n_estimators=1000, random_state=42,n_jobs=-1)
xgb_model.fit(X_train,y_train,early_stopping_rounds=10,eval_set=[(X_test, y_test)],verbose = 1)
y_pred = xgb_model.predict(X_test)

xgb_mae = mean_absolute_error(y_test, y_pred)
print(f'R2 Score: {round(xgb_model.score(X_test,y_test)*100,2)},\t MSE:{mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_pred=y_pred,y_true=y_test))}')
print(f"Mean Absolute Error: {round(xgb_mae, 4)}")
```

## Cross Validation
