### Holiday Parsing

```python
def holiday(holiday):
    cat = 'None'
    if holiday != 'None':
        cat = 'Holiday'
        
    return cat

df['holiday'] = df['holiday'].map(holiday)
df['holiday'].unique()
```

### Parsing do date_time

```python
def parse_timestamp(df, datetime):
    df[datetime] = pd.to_datetime(df[datetime])
    df['Year'] = df[datetime].dt.year
    df['Month'] = df[datetime].dt.month
    df['Weekday'] = df[datetime].dt.weekday
    df['Hour'] = df[datetime].dt.hour
    
def categorize_hour(hour):    
    cat = 'None'
    
    if hour in [1,2,3,4,5]:
        cat = 'dawn'
    elif hour in [6,7,8,9,10,11,12]:
        cat = 'morning'
    elif hour in [13,14,15,16,17,18]:
        cat = 'afternoon'
    elif hour in [19,20,21,22,23,24]:
        cat = 'night'
        
    return cat   
```

Antes de realizar o parsing de date_time irei remover possíveis duplicatas.

```python
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

## Importância de atributos

Irei rodar uma Random Forest a fim de encontrar a importância de cada atributo para o resultado de traffic_volume. Porém é necessário transformar os dados categóricos antes.

### Separando os atributos de X e y

```python
feats = df.drop('traffic_volume',1).copy()
label = df['traffic_volume'].copy()

feats.head()
```

### Fazendo uma cópia com Label Enconder de X

```python
le = LabelEncoder()
num_feats = feats.copy()

for column in num_feats.columns:
    if num_feats[column].dtype == 'object':
        num_feats[column] = le.fit_transform(num_feats[column])

num_feats.head()
```

### Fazendo uma cópia com One Hot Enconding de X

```python
onehot_feats = feats.copy()
onehot_feats[['Year','Month','Weekday']] = feats[['Year','Month','Weekday']].astype('category')
onehot_feats = pd.get_dummies(onehot_feats)

onehot_feats.head()
```

### Importância de atributos na estrutura após Label encoder

```python
model = RandomForestRegressor(random_state=1, max_depth=10)
model.fit(num_feats,label)

#Plot
features = feats.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-5:] #Top 5 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

### Importância de atributos na estrutura após One Hot Encoding

```python
model = RandomForestRegressor(random_state=1, max_depth=10)
model.fit(onehot_feats,label)
#Plot t
features = onehot_feats.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-10:] #Top 10 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

É claro então perceber que o horário é o atributo que mais influencia na quantidade do volume de carros. Isso pode ser observado empiricamente em pontos específicos de cidades com um grande número de carros. Em horários de pico o volume de carros é maior.

## Normalização

Necessária para manter os valores numéricos em um intervalo e assim um atributo não se destacar muito mais do que o outro na hora de treinar o modelo

```python
X = num_feats.values
X = preprocessing.scale(X)
y = label.values

print(X)
print(y)
```

## Treino e Teste

### Holdout(0.7; 0.3)

O conjunto de dados será divido aleatoriamente entre treino e teste. Para o problema em questão faremos a divisão entre 70% serem utilizados para treino, enquanto o resto será utilizado para teste.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state = 42)
X_train.shape
```

#### Random Forest

É um algoritmo de AM(Aprendizagem de máquina) supervisionado. Cria várias árvores de decisão e as combina para obter uma melhor predição.
<br/>
<br/>
*Random forest* adiciona aleatoriedade ao modelo quando está criando as árvores. Ele busca o melhor atributo em um subconjunto aleatório de atributo. Isso cria diversidade e a geração de modelos melhores.
Por essa razão foi possível encontrar a importância de atributos, realizada anteriormente.
<br/>
<br/>
<br/>
**Vantagens:**
1. Pode ser usado para classificação e regressão.
2. Lida bem com valores ausentes.
3. Dificilmente vai superajustar o modelo.
4. Lida muito bem com grandes conjuntos de dados e com alta dimensionalidade (Razão principal da escolha desse método)
<br/>
<br/>
**Desvantagens:**
1. O programador não possui muito controle no modelo, gera caixa preta
<br/>
<br/>
<br/>
**Exemplo de Random Forest:**
<div>
<img src="imgs/random_forest.png" width="500"/>
</div>

```python
reg = RandomForestRegressor()
reg = reg.fit(X_train,y_train)
predict = reg.predict(X_test)
accuracy = reg.score(X_test, y_test)
mae = mean_absolute_error(y_test, predict)

print(f"R2 Score: {round(accuracy * 100, 3)}%")
print(f"Mean Absolute Error: {round(mae, 4)}")
print(f'RMSE: {round(np.sqrt(mean_squared_error(y_pred=predict,y_true=y_test)),4)}')
```

### XGBoost

É uma implementação do algoritmo de AM supervisionado *Gradient Boost*, famoso por sua velocidade e performance. O *Gradient Boost* tenta prever uma variável de destino, para isso combina um grupo de estimativas de um conjunto de modelos mais simples e mais 'fracos' e os tranforma em um modelo mais 'forte'.
<br/>
<br/>
**Exemplo de Gradient Boosting:**
<div>
<img src="imgs/gradboost.png" width="350"/>
</div>

Nesse exemplo, é possível observar que o *Gradient Boosting* é um método em que novos modelos são treinados para resolver os erros de modelos anteriores.

```python
xgb_model = xgb.XGBRegressor(objective="reg:squarederror",  n_estimators=1000, random_state=42,n_jobs=-1)
xgb_model.fit(X_train,y_train,early_stopping_rounds=10,eval_set=[(X_test, y_test)],verbose = 0)
y_pred = xgb_model.predict(X_test)

xgb_mae = mean_absolute_error(y_test, y_pred)
print(f'R2 Score: {round(xgb_model.score(X_test,y_test)*100,2)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_pred=y_pred,y_true=y_test))}')
print(f"Mean Absolute Error: {round(xgb_mae, 4)}")
```

## Cross-Validation

O conjunto de dados é dividido em k partes, enquanto k-1 partes são utilizadas para treino a restante é utilizada para teste, até que todas as partes sejam utilizadas para treino e teste. A validação de cada previsão feita é salva em uma lista e depois será feita a média de cada métrica.

```python
kf = model_selection.KFold(n_splits=10, shuffle=True, random_state=42)
```

### Random Forest

```python
rmse = []
r2 = []
mae_l = []
for train_index, test_index in kf.split(X):
    reg_kf = RandomForestRegressor()
    reg_kf.fit(X[train_index],y[train_index])
    y_pred = reg_kf.predict(X[test_index])
    actuals = y[test_index]
    rmse.append(np.sqrt(mean_squared_error(actuals, y_pred)))
    r2.append(r2_score(actuals,y_pred))
    mae_l.append(mean_absolute_error(actuals, y_pred))


avg_r2 = np.mean(r2)
avg_rmse = np.mean(rmse)
avg_mae = np.mean(mae_l)
print(f'AVG R2 Score :{round(avg_r2,3)*100}%\t Average RMSE:{avg_rmse}\t Average MAE:{round(avg_mae, 4)}')    
```

### XGBoost

```python
rmse = []
r2 = []
mae_l = []
for train_index, test_index in kf.split(X):
    xgb_model = xgb.XGBRegressor(objective="reg:squarederror",  n_estimators=1000, random_state=42,n_jobs=-1)
    xgb_model.fit(X[train_index],y[train_index],early_stopping_rounds=10,eval_set=[(X[test_index], y[test_index])], verbose = 0)
    predictions = xgb_model.predict(X[test_index])
    actuals = y[test_index]
    rmse.append(np.sqrt(mean_squared_error(actuals, predictions)))
    r2.append(r2_score(actuals,predictions))
    mae_l.append(mean_absolute_error(actuals, predictions))

avg_r2 = np.mean(r2)
avg_rmse = np.mean(rmse)
avg_mae = np.mean(mae_l)
print(f'AVG R2 Score :{round(avg_r2,3)*100}%\t Average RMSE:{avg_rmse}\t Average MAE:{round(avg_mae, 4)}')
```

## Resultados e trabalhos futuros

As métricas escolhidas para a avaliação dos modelos foram:
1. RMSE(Root Mean Squared Error), consiste em encontrar a média da diferença entre o valor real e o previsto.
2. MAE(Mean Absolute Error), é calculada como a média do módulo da diferença entre o valor real e o previsto.
3. R²(R Squared), é uma medida que demonstra o quanto que o modelo explica a variância do valor real. Varia entre zero e um, quanto mais próximo de um mais ajustado à amostra está o modelo.



<p> A partir dos resultados apresentados para cada modelo é perceptível que as previsões se ajustam bem à amostra esperada. O método de Holdout apresentou melhor eficiência para o problema do que a validação cruzada, apesar dos resultados não apresentarem tantas diferenças. O algoritmo de XGBoost obteve previsões com menor erro e mais ajustadas à amostra original.
<br/>
<p> Apesar do satisfatório resultado dos modelos preditivos, um modelo que possui um RMSE com valores entre 390 e 400 pode não ser tão eficiente para determinadas aplicações. Por exemplo, um sistema de recomendação de rotas, em que o valor deve ser muito preciso.
Na finalidade de resolver esse problema, espera-se de trabalhos futuros a implementação de modelos específicos para problemas baseados em tempo, como por exemplo Recurrent Neural Networks(RNN) e ARIMA. Visto que os atributos que possuem maior relevância para o problema são relacionados a séries temporais.
